Replace the Missing values with median category wise

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

Clean columns name into lower casing and spaces with underscores

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a new column aged_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['aged_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','aged_group']].head(10)
```

```python
#Purchased frequency days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib

# 1. Your specific connection details from the image
# 'r' before the string handles the backslashes correctly
server = r'DESKTOP-BQDNLBO\SQLEXPRESS'
database = 'customer_behavior'
driver = 'ODBC Driver 17 for SQL Server'

# 2. Build the connection string for Windows Authentication
# We remove username/password and add 'Trusted_Connection=yes'
params = urllib.parse.quote_plus(
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
    f"TrustServerCertificate=yes;"
)

# 3. Create the engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 4. Write and Read back
try:
    df.to_sql("customer", engine, if_exists="replace", index=False)
    print("✅ Data successfully stored in 'customer' table.")
    
    # Verification query
    result = pd.read_sql("SELECT TOP 5 * FROM customer", engine)
    print(result)
except Exception as e:
    print(f"❌ Error: {e}")
```
